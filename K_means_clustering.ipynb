```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
%matplotlib inline

#Read in the image
image=cv2.imread('rolls-royce-cullinan.jpg')

#change colour to rgb(from bgr)
image=cv2.cvtColor(image,cv2.COLOR_BGR2RGB)

plt.imshow(image)
```

```python

#for applying kmeans convert it from 3d to 2d
#reshaping the img into a 2d array of pixels and 3 color values(RGB)
pixel_vals=image.reshape((-1,3))

#convert to float type
pixel_vals=np.float32(pixel_vals)
#k=6 increasing value of k makes it more clearer
#the below line of code defines the criteria for the algorithm to stop running
#which will happen in 100 iterations are run or the epsilon(which is req. acuracy)
#becomes 85%

criteria=(cv2.TERM_CRITERIA_EPS+cv2.TERM_CRITERIA_MAX_ITER, 100, 0.85)

#then perform kmeans clustering with the numbwer of clusters defines as 3
#also random centers are intially choosed for k means clustering

k=3
retval,labels,centers=cv2.kmeans(pixel_vals,k,None, criteria,10 , cv2.KMEANS_RANDOM_CENTERS)

#convert into 8 bit values
centers=np.uint8(centers)
segmented_data=centers[labels.flatten()]

#reshape data into original image dimensions
segmented_image=segmented_data.reshape((image.shape))


plt.imshow(segmented_image)

```

```python
import matplotlib.pyplot as plt

X=[4,5,10,4,3,11,14,6,10,12]
Y=[21,19,24,17,16,25,24,22,21,21]

plt.scatter(X,Y)
plt.show()

from sklearn.cluster import KMeans

data=list(zip(X,Y))
inertias=[]

for i in range(1,11):
  kmeans=KMeans(n_clusters=i)
  kmeans.fit(data)
  inertias.append(kmeans.inertia_)

plt.plot(range(1,11),inertias,marker='o')
plt.title('Elbow method')
plt.xlabel('Number of clusters')
plt.xlabel('Inertias')
plt.show()
```
